```python
import requests
import pandas as pd
import json
from datetime import datetime

site = requests.get('https://api.github.com/repos/thisismetis/dsp/stats/commit_activity')
data1 = site.json()
print(data1)
df = pd.DataFrame.from_dict(data1)
```

```python
print(df)
```

```python
import matplotlib.pyplot as plt
df['date'] = pd.to_datetime(df.week, unit='s')
df['week_no'] = df.apply(lambda x: '{:d}-{:02d}'.format(x['date'].year, x['date'].weekofyear), axis=1)
df.set_index('week_no')['total'].plot.bar()
```
