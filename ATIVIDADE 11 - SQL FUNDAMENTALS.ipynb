```python
#Criando um cursor para executar consultas SQL em um banco de dados
cur = conn.cursor()

#Pegando o nome das tabelas dentro do banco de dados
tables = cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
for name in tables:
    print (name[0])
```
